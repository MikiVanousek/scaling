```python
import datasets
from tqdm import tqdm

ID = 'Tevatron/wiki-ss-corpus'
ds = datasets.load_dataset(ID, split='train[0:1000000]', column_names=['text'] )
ds
```

```python
characters = set('abcdefghijklmnopqrstuvwxyz!@#$%^&*(),.!?0123457890"\'')
from collections import Counter, defaultdict
char_count = Counter()
token_count = Counter()

for row in tqdm(ds):
    text = row['text'].lower()
    char_count.update(list(text))
    token_count.update(text.split(' '))

    

```

```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace
from datasets import load_dataset

dataset = ds

# Create an empty tokenizer
tokenizer = Tokenizer(BPE(unk_token="[UNK]"))

# Optional: use GPT-2 style byte-level pre-tokenization
# tokenizer.pre_tokenizer = ByteLevel()
tokenizer.pre_tokenizer = Whitespace()

trainer = BpeTrainer(
    vocab_size=100000,         # temporary value (you can change this later)
    min_frequency=2,
    special_tokens=["[PAD]", "[UNK]", "[BOS]", "[EOS]"]
)

def batch_iterator():
    for batch in ds:
        yield batch["text"]

tokenizer.train_from_iterator(batch_iterator(), trainer=trainer)

tokenizer.save("bpe-tokenizer.json")
```

```python
from transformers import PreTrainedTokenizerFast
from rich import print

# Load your tokenizer
tok = PreTrainedTokenizerFast(tokenizer_file="bpe-tokenizer.json")

def show_tokens(text):
    out = tok.encode_plus(text, return_offsets_mapping=True)
    ids = out["input_ids"]
    offsets = out["offset_mapping"]

    print("[bold underline]Tokenization:[/]\n")
    for token_id, (start, end) in zip(ids, offsets):
        token = text[start:end]
        piece = tok.convert_ids_to_tokens([token_id])[0]
        print(f"[cyan]{piece:<15}[/]    '{token}'    (id={token_id})")
show_tokens(dataset[1]['text'])
```

```python
# Add a new column with tokenized length
import numpy as np

def add_length(example):
    example["len"] = len(tok.encode(example["text"]))
    return example

ds = ds.map(add_length)

# Extract lengths as a NumPy array
lengths = np.array(ds["len"])

# Compute quantiles
quantiles = [0.1 * i for i in range(1, 10)]
values = np.quantile(lengths, quantiles)

for q, v in zip(quantiles, values):
    print(f"{int(q*100):2d}th percentile: {v:.0f}")
```
