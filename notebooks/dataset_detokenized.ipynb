```python
import pandas as pd
from datasets import load_dataset
from transformers import AutoTokenizer

# 1. Load the dataset (streaming is faster for previewing large datasets)
dataset = load_dataset("MikiV/Ultra-FineWeb-chunked-4x2048", split="train", streaming=True)

# 2. Initialize the tokenizer (FineWeb uses the GPT-2/Llama-style byte-level BPE)
tokenizer = AutoTokenizer.from_pretrained("gpt2")

# 3. Take a sample (e.g., first 5 rows) and detokenize
data_samples = []
for example in dataset:
    # 'input_ids' is the standard key for tokenized data in this dataset
    decoded_text = tokenizer.decode(example['input_ids'], skip_special_tokens=True)
    data_samples.append({"decoded_content": decoded_text})

# 4. Create DataFrame and display
df = pd.DataFrame(data_samples)
display(df)
```
